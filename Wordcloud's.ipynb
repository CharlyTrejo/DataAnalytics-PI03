# Wordcloud's

Este notebook tiene como objetivo generar Wordclouds de las palabras más repetidas dentro de los títulos de los cursos, 

```
import nltk
import pandas as pd
import matplotlib.pyplot as plt
nltk.download('stopwords')
from wordcloud import WordCloud 
from nltk.corpus import stopwords
stopwords = stopwords.words('english')
```

Ingesto en dataframes los datasets anteriormente transformados

```
df_coursera_courses = pd.read_csv(r'coursera-courses.csv', sep=',', header=0)
df_coursera_reviews = pd.read_csv(r'coursera-review.csv', sep=',', header=0)
df_edx = pd.read_csv(r'edx.csv', sep=',', header=0)
df_udemy = pd.read_csv(r'udemy.csv', sep=',', header=0)
```

Separo en objetos Series el campo de interés para el Wordcloud

```
title_coursera = df_coursera_courses['name']
title_edx = df_edx['title']
title_udemy = df_udemy['course_title']
```

## Wordcloud Coursera 'name'

```
description_plain_text = title_coursera.str.cat(sep=' ')
cloud_prueba = WordCloud(stopwords = stopwords).generate(description_plain_text)
plt.figure(figsize=(20,10))
plt.imshow(cloud_prueba, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## Wordcloud edx 'title'

```
description_plain_text = title_edx.str.cat(sep=' ')
cloud_prueba = WordCloud(stopwords = stopwords).generate(description_plain_text)
plt.figure(figsize=(20,10))
plt.imshow(cloud_prueba, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## Wordcloud udemy 'course_title'

```
description_plain_text = title_udemy.str.cat(sep=' ')
cloud_prueba = WordCloud(stopwords = stopwords).generate(description_plain_text)
plt.figure(figsize=(20,10))
plt.imshow(cloud_prueba, interpolation='bilinear')
plt.axis('off')
plt.show()
```

## Wordcloud global

Ahora generaré un wordcloud que incluya todos los títulos

```
title_global = pd.concat([title_coursera, title_edx, title_udemy])
title_global
```

```
description_plain_text = title_global.str.cat(sep=' ')
cloud_prueba = WordCloud(stopwords = stopwords).generate(description_plain_text)
plt.figure(figsize=(20,10))
plt.imshow(cloud_prueba, interpolation='bilinear')
plt.axis('off')
plt.show()
```
