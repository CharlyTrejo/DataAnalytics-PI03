## Registros duplicados

Analizo si hay registros duplicados y los elimino; aunque probablemente existan valores duplicados en campos como la descripción, y reviews, no me es importante eliminarlos ya que puede haber comentarios similares con palabras como "bad", "good", "excelent", etc. Por otra parte eliminaría insights importantes como personas inscritas en un curso, mismos datos que serán importantes al hacer la segmentación. 

### Coursera

```
df_coursera_courses.duplicated().sum()
```

```
df_coursera_reviews.duplicated().sum()
```

Procedo a eliminar registros duplicados : 

```
df_coursera_reviews.drop_duplicates(inplace=True)
```

Confirmo la eliminación de mis registros duplicados

```
print(df_coursera_reviews.duplicated().sum())
print(df_coursera_reviews.shape)
```

### edX

```
df_edx_courses.duplicated().sum()
```

Elimino registros duplicados: 

```
df_edx_courses.drop_duplicates(inplace=True)
```

Confirmo eliminación de registros duplicados 

```
print(df_edx_courses.duplicated().sum())
print(df_edx_courses.shape) 
```

### Udemy

```
df_udemy_courses.duplicated().sum()
```
